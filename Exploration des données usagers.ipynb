```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
usagers = pd.read_csv("Jeux_de_données/usagers-2019.csv", delimiter=";")
usagers
```

```python
usagers.info()
```

```python
usagers.replace(-1, np.nan, inplace=True)
usagers
```

```python
usagers = usagers.replace(" -1", np.NAN)
usagers = usagers.replace("0", np.NAN)
usagers['trajet'] = usagers['trajet'].replace(0, np.NAN)
usagers
```

```python
usagers.isnull().sum()
```

```python
for i in usagers.columns:
    print("TYPE : ",usagers[i].dtype)
    print("Value counts : ", usagers[i].value_counts())
```

```python
usagers_sans_id_unique = usagers[["num_veh", "place", "catu", "grav", "sexe", "an_nais", "trajet", "secu1", "secu2", "secu3", "locp", "actp", "etatp"]]
```

```python
sns.set(style="white", color_codes=True)
sns.pairplot(data=usagers_sans_id_unique)
plt.show()
```

```python
correlation = usagers_sans_id_unique.corr()
plt.figure(figsize=(15, 8))
plt.subplot(111)
sns.heatmap(correlation, annot=True, linewidths=1.0)
plt.show()
```

```python
cat_usager = usagers_sans_id_unique['catu'].value_counts().keys()
donnee_bar = usagers_sans_id_unique.loc[usagers_sans_id_unique['catu'].isin(cat_usager)].groupby(['catu', 'sexe']).size().unstack(1)

valeurs_hommes = donnee_bar[1]
valeurs_femmes = donnee_bar[2]

width = 0.35
plt.figure()
plt.subplot(111)

plt.bar(donnee_bar.index, valeurs_hommes, width, label='Hommes', color='blue')
plt.bar(donnee_bar.index, valeurs_femmes, width, bottom=valeurs_hommes, label='Femmes', color="red")

plt.ylabel('Nombre')
plt.xlabel("Nombre d'hommes ou de femmes par catégories")

plt.xticks(donnee_bar.index)
plt.legend()
plt.show()
```

```python
grav_usager = usagers_sans_id_unique['grav'].value_counts().keys()
donnee_bar = usagers_sans_id_unique.loc[usagers_sans_id_unique['grav'].isin(grav_usager)].groupby(['grav', 'sexe']).size().unstack(1)

valeurs_hommes = donnee_bar[1]
valeurs_femmes = donnee_bar[2]

width = 0.35
plt.figure()
plt.subplot(111)

plt.bar(donnee_bar.index, valeurs_hommes, width, label='Hommes', color='blue')
plt.bar(donnee_bar.index, valeurs_femmes, width, bottom=valeurs_hommes, label='Femmes', color="red")

plt.ylabel('Nombre')
plt.xlabel("Nombre d'hommes ou de femmes par gravité d'accident")

plt.xticks(donnee_bar.index)
plt.legend()
plt.show()
```

```python
palette= sns.color_palette("RdBu", 10)
sns.set_palette(palette)
sns.set_style("white")

plt.figure(figsize=[10,6])

plt.subplot(111)
sns.scatterplot(data=usagers_sans_id_unique, hue="sexe",
                x='an_nais', y='trajet')
plt.title("---")

plt.show()
```

```python
grav_usager = usagers_sans_id_unique['grav'].value_counts().keys()
donnee_bar = usagers_sans_id_unique.loc[usagers_sans_id_unique['grav'].isin(grav_usager)].groupby(['grav', 'locp']).size().unstack(1)

valeurs_sans_objet = donnee_bar[0]
valeurs_chaussee_plus = donnee_bar[1]
valeurs_chaussee_moins = donnee_bar[2]
valeur_passage_pieton_sans = donnee_bar[3]
valeur_passage_pieton_avec = donnee_bar[4]
valeur_divers_1 = donnee_bar[5]
valeur_divers_2 = donnee_bar[6]
valeur_divers_3 = donnee_bar[7]
valeur_divers_4 = donnee_bar[8]
valeur_divers_5 = donnee_bar[9]

width = 0.35
plt.figure(figsize=(18,8))
plt.subplot(111)


plt.bar(donnee_bar.index, valeurs_sans_objet, width, label='Sans objet', color='grey')
plt.bar(donnee_bar.index, valeurs_chaussee_plus, width, bottom=valeurs_sans_objet, label='Sur la chaussée + 50m du passage piéton', color='red')
plt.bar(donnee_bar.index, valeurs_chaussee_moins, width, bottom=valeurs_chaussee_plus, label='Sur la chaussée - 50m du passage piéton', color='orange')
plt.bar(donnee_bar.index, valeur_passage_pieton_sans, width, bottom=valeurs_chaussee_moins, label='Passage piéton sans signalisation lumineuse', color='yellow')
plt.bar(donnee_bar.index, valeur_passage_pieton_avec, width, bottom=valeur_passage_pieton_sans, label='Passage piéton avec signalisation lumineuse', color='green')
plt.bar(donnee_bar.index, valeur_divers_1, width, bottom=valeur_passage_pieton_avec, label='Sur trottoir', color='#5700A7')
plt.bar(donnee_bar.index, valeur_divers_2, width, bottom=valeur_divers_1, label='Sur accotement', color='#5000AE')
plt.bar(donnee_bar.index, valeur_divers_3, width, bottom=valeur_divers_2, label='Sur refuge ou BAU', color='#4A00B4')
plt.bar(donnee_bar.index, valeur_divers_4, width, bottom=valeur_divers_3, label='Sur contre allée', color='#4400BA')
plt.bar(donnee_bar.index, valeur_divers_5, width, bottom=valeur_divers_4, label='Inconnue', color='#3E00C0')

plt.ylabel('Nombre')
plt.xlabel("localisation du piétion par gravité d'accident")

plt.xticks(donnee_bar.index)
plt.legend()
plt.show()
```

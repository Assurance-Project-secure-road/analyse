```python
import pandas as pd
import json
from sqlalchemy.engine import create_engine
```

```python
## IMPORT DU CONFIG.JSON
# assignation de la config.json à fichierConfig
fichierConfig = "config.json"
# ouverture et chargement des donnée contenu dans fichierConfig
with open(fichierConfig) as fichier:
    config = json.load(fichier)["Postgres"]


class SqlORM:
    def __init__(self, config):
        self.config = config
        self.connector = self._connect_db()

    def _connect_db(self):
        connector = create_engine(
            config["connector"]
            + "://"
            + config["user"]
            + ":"
            + config["password"]
            + "@"
            + config["host"]
            + ":"
            + config["port"]
            + "/"
            + config["bdd"],
            echo=False,
        )
        return connector


testclass = SqlORM(config)
print(10 * "*")
print("test de la connection", "\n")
connection = testclass.connector
print(connection, "\n")

```

```python
from sqlalchemy import Column, Integer, VARCHAR, DateTime, DECIMAL, ForeignKey
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.sql import func


Base = declarative_base()

class Caracteristique(Base):
    __tablename__ = 'Caracteristique'
    Num_Acc_id = Column('Num_Acc_id', Integer, primary_key=True)
    Date_Acc = Column('Date_Acc', DateTime)
    Lumiere_Acc = Column('Lumiere_Acc', Integer)
    Departement_Acc = Column("Departement_Acc", VARCHAR)
    Commune_Acc = Column("Commune_Acc",VARCHAR)
    Agglomeration_Acc = Column("Agglomeration_Acc", Integer)
    Intersection_Acc = Column("Intersection_Acc", Integer)
    Meteo_Acc = Column("Meteo", Integer)
    Collision_Acc = Column("Collision_Acc",Integer)
    Addresse_Acc = Column("Addresse_Acc", VARCHAR)
    Latitude_Acc = Column("Latitude_Acc", DECIMAL(10,7))
    Longitude_Acc = Column("Longitude_Acc",DECIMAL(10,7))
    

class Vehicule(Base):
    __tablename__ = 'Vehicule'
    id_Vehicule = Column('id_vehicule', Integer, primary_key=True)
    Num_Vehicule = Column('Num_Vehicule', Integer)
    Sens_circulation_Vehicule = Column("Sens_circulation_Vehicule", Integer)
    Categorie_Vehicule = Column("Categorie_Vehicule", Integer)
    Obstacle_fixe_heurete_Vehicule = Column("Obstacle_fixe_heurete_Vehicule", Integer)
    Obstacle_mobile_heurete_Vehicule = Column("Obstacle_mobile_heurete_Vehicule", Integer)
    Point_choc_Vehicule = Column("Point_choc_Vehicule",Integer)
    Manoeuvre_avant_accident_Vehicule = Column("Manoeuvre_avant_accident_Vehicule", Integer)
    Type_motorisation_Vehicule = Column("Type_motorisation_Vehicule", Integer)
    Nb_occupant_Vehicule = Column("Nb_occupant_Vehicule",Integer)
    Num_Acc_id = Column("Num_Acc_id",ForeignKey("Caracteristique.Num_Acc_id"))


class Usager(Base):
    __tablename__ = 'Usager'
    id_Usager = Column('id_Usager', Integer, primary_key=True)
    Place_Usager = Column('Place_Usager', Integer)
    Categorie_Usager = Column("Categorie_Usager", Integer)
    Gravblessure_Usager = Column("Gravblessure_Usager", Integer)
    Sexe_Usager = Column("Sexe_Usager", Integer)
    Anne_Naissance_Usager = Column("Anne_Naissance_Usager", Integer)
    Motif_Deplacer_Usager = Column("Motif_Deplacer_Usager",Integer)
    Equipement_Secu_Usager = Column("Equipement_Secu_Usager", Integer)
    Localisation_pieton_Usager = Column("Localisation_pieton_Usager", Integer)
    Action_pieton_Usager = Column("Action_pieton_Usager", VARCHAR)
    Nb_pieton_Usager = Column("Nb_pieton_Usager",Integer)
    Num_Acc_id = Column("Num_Acc_id",ForeignKey("Caracteristique.Num_Acc_id"))
    id_Vehicule = Column("id_Vehicule", ForeignKey("Vehicule.id_vehicule"))


class Lieux(Base):
    __tablename__ = 'Lieux'
    ID_Route = Column("ID_Route", Integer, primary_key=True)
    Categorie_Route = Column('Categorie_Route', Integer)
    Numero_Route = Column('Numero_Route', Integer)
    Indice_num_Route = Column('Indice_num_Route', Integer)
    Indice_alphanum_Route = Column("Indice_alphanum_Route", VARCHAR(5))
    Circulation_Route = Column("Circulation_Route", Integer)
    Nb_voie_Route = Column("Nb_voie_Route",Integer)
    Type_voie_Route = Column("Type_voie_Route", Integer)
    Declivite_Route = Column("Declivite_Route", Integer)
    Num_borne_Route = Column("Num_borne_Route", Integer)
    Distance_borne_Route = Column("Distance_borne_Route", Integer)
    Tracer_plan_Route = Column("Tracer_plan_Route", Integer)
    Largeur_TPC_Route = Column("Largeur_TPC_Route", Integer)
    Largeur_Route = Column("Largeur_Route", Integer)
    Surface_Route = Column("Surface_Route", Integer)    
    Infrastruture_Route = Column("Infrastruture_Route", Integer)
    Situation_Acc_Route = Column("Situation_Acc_Route", Integer)
    Vitesse_max_Route = Column("Vitesse_max_Route", Integer)
    Num_Acc_id = Column("Num_Acc_id",ForeignKey("Caracteristique.Num_Acc_id"))
```

```python
Base.metadata.create_all(connection)
```

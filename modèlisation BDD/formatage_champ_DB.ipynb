```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime


data_Lieux = pd.read_csv("/home/virus/Projet assurance sécurité routière/analyse/lieux-2019.csv", delimiter=";")


data_Caracteristiques = pd.read_csv("/home/virus/Projet assurance sécurité routière/analyse/caracteristiques-2019.csv", delimiter=";", decimal=",")


data_Usagers = pd.read_csv("/home/virus/Projet assurance sécurité routière/analyse/usagers-2019.csv", delimiter=";", decimal=",")


data_vehicules = pd.read_csv("/home/virus/Projet assurance sécurité routière/analyse/vehicules-2019.csv", delimiter=";", decimal=",")
```

```python
def Format_Dataframe_Lieux(data_Lieux):

    data_Lieux["catr"].replace(9, np.nan, inplace=True)
    data_Lieux["circ"].replace(-1, np.nan, inplace=True)
    data_Lieux["vosp"].replace(-1, np.nan, inplace=True)
    data_Lieux["prof"].replace(-1, np.nan, inplace=True)
    data_Lieux["vma"].replace(-1, np.nan, inplace=True)
    data_Lieux["prof"].replace(-1, np.nan, inplace=True)
    data_Lieux["surf"].replace([-1, 9], np.nan, inplace=True)
    data_Lieux["infra"].replace([-1, 9], np.nan, inplace=True)
    data_Lieux["situ"].replace([-1, 8], np.nan, inplace=True)

    data_Lieux = data_Lieux.rename(
        columns={
            "Num_Acc": "Num_Acc_id",
            "catr": "Categorie_Route",
            "voie": "Numero_Route",
            "v1": "Indice_num_Route",
            "v2": "Indice_alphanum_Route",
            "circ": "Circulation_Route",
            "nbv": "Nb_voie_Route",
            "vosp": "Type_voie_Route",
            "prof": "Declivite_Route",
            "pr": "Num_borne_Route",
            "pr1": "Distance_Borne_Route",
            "plan": "Tracer_plan_Route",
            "lartpc": "Largeur_TPC_Route",
            "larrout": "Largeur_Route",
            "surf": "Surface_Route",
            "infra": "Infrastructure_Route",
            "situ": "Situation_Acc_Route",
            "vma": "Vitesse_max_Route",
        }
    )

    data_Lieux = data_Lieux.reset_index()
    data_Lieux = data_Lieux.rename(columns={"index": "id_Route"})
    data_Lieux.id_Route = data_Lieux.id_Route + 1

    col_name = "Num_Acc_id"
    fk_col = data_Lieux.pop(col_name)
    data_Lieux.insert(18, col_name, fk_col)

    return data_Lieux
```

```python
Format_Dataframe_Lieux(data_Lieux)
```

```python
def Format_Dataframe_Caracteristiques(data_Caracteristiques):
    
    data_Caracteristiques["atm"] = data_Caracteristiques["atm"].apply(lambda x: pd.NA if x == -1 else x)
    data_Caracteristiques["col"] = data_Caracteristiques["col"].apply(lambda x: pd.NA if x == -1 else x)
    data_Caracteristiques["date"] = data_Caracteristiques.apply(lambda line: datetime(line["an"], line["mois"], line["jour"], int(line["hrmn"].split(":")[0]), int(line["hrmn"].split(":")[1])), axis=1)



    
    data_Caracteristiques = data_Caracteristiques.rename(
        columns={
            "Num_Acc":"Num_Acc_id",
            "date":"Date_Acc",
            "lum":"Lumiere_Acc",
            "dep":"Departement_Acc",
            "com":"Commune_Acc",
            "agg":"Agglomeration_Acc",
            "int":"Intersection_Acc",
            "atm":"Meteo_Acc",
            "col":"Collision_Acc",
            "adr":"Addresse_Acc",
            "lat":"Latitude_Acc",
            "long":"Longitude_Acc"
            
        })
    
    data_Caracteristiques = data_Caracteristiques.drop(columns={"jour","mois","an","hrmn"})

    
    return data_Caracteristiques
```

```python
Format_Dataframe_Caracteristiques(data_Caracteristiques)
```

```python
def Format_Dataframe_Usagers(data_Usagers):
    data_Usagers.replace(-1, np.nan, inplace=True)
    data_Usagers = data_Usagers.replace(" -1", np.NAN)
    data_Usagers = data_Usagers.replace("0", np.NAN)
    data_Usagers["trajet"] = data_Usagers["trajet"].replace(0, np.NAN)
    data_Usagers = data_Usagers.rename(
        columns={
            "Num_Acc": "Num_Acc_id",
            "place": "Place_Usager",
            "catu": "Categorie_Usager",
            "grav": "Gravblessure_Usager",
            "sexe": "Sexe_Usager",
            "an_nais": "Anne_Naissance_Usager",
            "trajet": "Motif_Deplacer_Usager",
            "secu1": "Equipement_Secu_Usager",
            "locp": "Localisation_pieton_Usager",
            "actp": "Action_pieton_Usager",
            "etatp": "Nb_pieton_Usager",
        }
    )


    data_Usagers = data_Usagers.reset_index()
    data_Usagers = data_Usagers.rename(columns={"index": "id_Usager"})
    data_Usagers.id_Usager = data_Usagers.id_Usager + 1
    
    
    data_Usagers = data_Usagers.drop(["num_veh", "secu2", "secu3"], axis=1)
    return data_Usagers
```

```python
Format_Dataframe_Usagers(data_Usagers)
```

```python
def Format_Dataframe_vehicules(data_vehicules):
    #recupère des cols pour affecter la valeur -1 à NaN
    col_replace_null =  data_vehicules.columns[3:-1]

    for col in col_replace_null:
        data_vehicules[col].replace(-1, np.nan, inplace=True)
        data_vehicules[col] = data_vehicules[col].astype("Int64")

    

    
    data_vehicules = data_vehicules.rename(
        columns={
            "Num_Acc": "Num_Acc_id",
            "id_vehicule": "id_Vehicule",
            "num_veh": "Num_Vehicule",
            "senc": "Sens_circulation_Vehicule",
            "catv": "Catégorie_Vehicule",
            "obs": "Obstacle_fixe_heurte_Vehicule",
            "obsm": "Obstacle_mobile_heurte_Vehicule",
            "choc": "Point_choc_Vehicule",
            "manv": "Manoeuvre_avant_accident_Vehicule",
            "motor": "Type_motorisation_Vehicule",
            "occutc": "Nb_occupant_Vehicule"
            
        }
    )

    return data_vehicules
```

```python
Format_Dataframe_vehicules(data_vehicules)
```
